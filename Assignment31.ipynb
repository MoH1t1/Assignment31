```python
# Q1: What is Estimation Statistics? Explain point estimate and interval estimate.

Estimation: A method of inferring the population parameter using sample data.
Point Estimate: A single value estimate of the population parameter (e.g., sample mean).
Interval Estimate: A range of values, with a level of confidence, that likely contains the population parameter (e.g., confidence interval).
```

```python
# Q2. Write a Python function to estimate the population mean using a sample mean and standard deviation.

import numpy as np

def estimate_population_mean(sample_data):
    return np.mean(sample_data)

sample = [50, 55, 60, 65, 70]
print(estimate_population_mean(sample))
```

```python
# Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.

A statistical method to test if a hypothesis about a population parameter is true.
Purpose: To make inferences about a population based on sample data.
Importance: Helps to validate assumptions and make decisions based on data.
```

```python
# Q4. Create a hypothesis that states whether the average weight of male college students is greater than the average weight of female college students.

Null Hypothesis (H₀): There is no difference in the average weight of male and female college students, i.e., the average weight of male students is equal to the average weight of female students.
H0: μmale =μfemale

Alternative Hypothesis (H₁): The average weight of male college students is greater than the average weight of female college students.
H1 : μmale > μfemale

 This is a one-tailed hypothesis test where we are testing if the average weight of male college students is greater than the average weight of female college students.
```

```python
# Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,given a sample from each population.

from scipy import stats

def hypothesis_test(sample1, sample2):
    t_stat, p_val = stats.ttest_ind(sample1, sample2)
    return t_stat, p_val

sample1 = [55, 60, 58, 62]
sample2 = [50, 53, 52, 51]
print(hypothesis_test(sample1, sample2))
```

```python
# Q6: What is a null and alternative hypothesis? Give some examples.

Null Hypothesis (H₀): No effect or difference. Example: The average height of men and women is the same.
Alternative Hypothesis (H₁): There is an effect or difference. Example: The average height of men is different from that of women.
```

```python
# Q7: Write down the steps involved in hypothesis testing.

Steps:
State the null and alternative hypotheses.
Choose a significance level (α).
Select the appropriate test and calculate the test statistic.
Find the p-value and compare with α.
Make a decision to reject or fail to reject the null hypothesis.
```

```python
# Q8. Define p-value and explain its significance in hypothesis testing.

Definition: The probability of obtaining results at least as extreme as those observed, assuming the null hypothesis is true.
Significance: A smaller p-value indicates stronger evidence against the null hypothesis.
```

```python
# Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
# random samples of equal size and a null hypothesis that the population means are equal.

from scipy import stats

def two_sample_t_test(sample1, sample2):
    t_stat, p_val = stats.ttest_ind(sample1, sample2)
    return t_stat, p_val

sample1 = [30, 32, 35, 33]
sample2 = [28, 29, 31, 30]
print(two_sample_t_test(sample1, sample2))
```

```python
# Q11: What is Student’s t distribution? When to use the t-Distribution.

Definition: A probability distribution used for small sample sizes, typically when population variance is unknown.
When to use: For hypothesis testing when sample size is small (less than 30) and the population standard deviation is unknown.
```

```python
# Q12: What is t-statistic? State the formula for t-statistic.

The t-statistic is a value used in hypothesis testing to determine if there is a significant difference between the sample mean and the population mean (or between two sample means). 

Formula:
For a single sample: t = (x - μ)/ s/root over(n)

Where:

x = Sample mean
μ = Population mean (or hypothesized mean)
s = Sample standard deviation
n = Sample size
```

```python
# Q13. A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random sample of 50 days and find the sample mean revenue to 
# be $500 with a standard deviation of $50. Estimate the population mean revenue with a 95% confidence interval.

import scipy.stats as stats
import numpy as np

sample_mean = 500
sample_std = 50
sample_size = 50
confidence_level = 0.95

error_margin = stats.t.ppf(1 - (1 - confidence_level) / 2, sample_size - 1) * (sample_std / np.sqrt(sample_size))
confidence_interval = (sample_mean - error_margin, sample_mean + error_margin)
print(confidence_interval)
```

```python
# Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a clinical trial with 100 patients and find that the 
# sample mean decrease in blood pressure is 8 mmHg with a standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.

from scipy import stats

sample_mean = 8
population_mean = 10
std_dev = 3
sample_size = 100
alpha = 0.05

t_stat = (sample_mean - population_mean) / (std_dev / (sample_size ** 0.5))
df = sample_size - 1
p_val = stats.t.cdf(t_stat, df)
print(f"t-statistic: {t_stat}, p-value: {p_val}")
```

```python
# Q15. An electronics company produces a certain type of product with a mean weight of 5 pounds and a
# standard deviation of 0.5 pounds. A random sample of 25 products is taken, and the sample mean weight
# is found to be 4.8 pounds. Test the hypothesis that the true mean weight of the products is less than 5 pounds with a significance level of 0.01.

from scipy import stats

sample_mean = 4.8
population_mean = 5
std_dev = 0.5
sample_size = 25
alpha = 0.01

t_stat = (sample_mean - population_mean) / (std_dev / (sample_size ** 0.5))
df = sample_size - 1
p_val = stats.t.cdf(t_stat, df)
print(f"t-statistic: {t_stat}, p-value: {p_val}")
```

```python
# Q16. Two groups of students are given different study materials to prepare for a test. The first group (n1 = 30) has a mean score of 80 with a standard 
# deviation of 10, and the second group (n2 = 40) has a mean score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two
# groups are equal with a significance level of 0.01.

import numpy as np
from scipy import stats

sample1 = [80] * 30
sample2 = [75] * 40
t_stat, p_val = stats.ttest_ind(sample1, sample2)
print(f"t-statistic: {t_stat}, p-value: {p_val}")
alpha = 0.01

if p_val < alpha:
    print("Reject the null hypothesis: The population means are different.")
else:
    print("Fail to reject the null hypothesis: The population means are equal.")
```

```python
# Q17. A marketing company wants to estimate the average number of ads watched by viewers during a TV program. They take a random sample of 50 viewers and find 
# that the sample mean is 4 with a standard deviation of 1.5. Estimate the population mean with a 99% confidence interval.

import numpy as np
import scipy.stats as stats

sample_mean = 4
sample_std = 1.5
sample_size = 50
confidence_level = 0.99

error_margin = stats.t.ppf(1 - (1 - confidence_level) / 2, sample_size - 1) * (sample_std / np.sqrt(sample_size))

confidence_interval = (sample_mean - error_margin, sample_mean + error_margin)

print(f"99% Confidence interval for population mean: {confidence_interval}")
```
